1. Даны значения зарплат из выборки выпускников: 100, 80, 75, 77, 89, 33, 45, 25, 65, 17, 30, 24, 57, 55, 70, 75, 65, 84, 90, 150. Посчитать (желательно без использования статистических методов наподобие std, var, mean) среднее арифметическое, среднее квадратичное отклонение, смещенную и несмещенную оценки дисперсий для данной выборки.

```python
array = [100, 80, 75, 77, 89, 33, 45, 25, 65, 17, 30, 24, 57, 55, 70, 75, 65, 84, 90, 150]
```

```python
def mean_(array):
    return sum(array)/len(array)

def var_(array):
    return mean_([(i-mean_(array))**2 for i in array])

def var_smpl(array):
    return sum((i-mean_(array))**2 for i in array)/(len(array)-1)

def std_(array):
    return var_(array)**0.5    
```

```python
print('среднее арифметическое:\n{:.1f}\nсреднее квадратичное отклонение:\n{:.1f}'
     '\nсмещенная дисперсия:\n{:.1f}\nнесмещенная дисперсия:\n{:.1f}'.format(mean_(array), std_(array),
                                                                           var_smpl(array), var_(array)))
```

2. В первом ящике находится 8 мячей, из которых 5 - белые. Во втором ящике - 12 мячей, из которых 5 белых. Из первого ящика вытаскивают случайным образом два мяча, из второго - 4. Какова вероятность того, что 3 мяча белые?

```python
# возможные комбинации белых
# ящик     1    2
# комб_0   0    3
# комб_1   1    2
# комб_2   2    1
```

```python
from math import factorial

def combinations(n, k):
    return int(factorial(n) / (factorial(k) * factorial(n-k)))
```

```python
# ящик 1: достать 2 из 3 небелых, ящик 2: 3 из 5 белых
p1 = (combinations(3, 2) / combinations(8, 2)) * combinations(5, 3) / combinations(12, 4)
# ящик 1: достать 1 из 5 белых, ящик 2: 2 из 5 белых
p2 = (combinations(5, 1) / combinations(8, 2)) * combinations(5, 2) / combinations(12, 4)
# ящик 1: достать 2 из 5 белых, ящик 2: 1 из 5 белых
p3 = (combinations(5, 2) / combinations(8, 2)) * combinations(5, 1) / combinations(12, 4)
print('вероятность достать 3 белых мяча составляет: {:.2%}'.format(p1+p2+p3))
```

3. На соревновании по биатлону один из трех спортсменов стреляет и попадает в мишень. Вероятность попадания для первого спортсмена равна 0.9, для второго — 0.8, для третьего — 0.6. Найти вероятность того, что выстрел произведен: a). первым спортсменом б). вторым спортсменом в). третьим спортсменом.

решать по формуле байеса 

$$ P(A|B) = \frac{P(A)*P(B|A)}{P(B)} $$

P(A) - вероятность того, что попал выбранный спортсмен

P(B|A) - априорная вероятность попадания выбранным спортсменом

P(B) - суммарная вероятность попадания

при этом, очевидно, что P(A) - определяется участников $ P(A)_1 == P(A)_2 == P(A)_3 == \frac{1}{3} $

```python
# вероятность, что какой то спортсмен попал
p_all = 1 / 3 * 0.9 + 1 / 3 * 0.8 + 1 / 3 * 0.6
# конкретно для факультетов
p1 = (1 / 3 * 0.9 ) / p_all
p2 = (1 / 3 * 0.8) / p_all
p3 = (1 / 3 * 0.6) / p_all

print('а) вероятность попадания первым спортменом: {:.2%}'.format(p1))
print('б) вероятность попадания вторым спортменом: {:.2%}'.format(p2))
print('в) вероятность попадания третьим спортменом: {:.2%}'.format(p3))
```

при это я захотел проверить другую формулировку задачи, и не могу точно сказать, почему следующая трактовка неверна:

задача формулируется как одновременное наступление событий - попадания одного из стрелков и непопаданания двух других:

```python
p1 = 0.9
p2 = 0.8
p3 = 0.6

print('а) вероятность попадания первым спортменом: {:.2%}'.format(p1*(1-p2)*(1-p3)))
print('б) вероятность попадания вторым спортменом: {:.2%}'.format((1-p1)*p2*(1-p3)))
print('а) вероятность попадания третьим спортменом: {:.2%}'.format((1-p1)*(1-p2)*p3))
```

4. В университет на факультеты A и B поступило равное количество студентов, а на факультет C студентов поступило столько же, сколько на A и B вместе. Вероятность того, что студент факультета A сдаст первую сессию, равна 0.8. Для студента факультета B эта вероятность равна 0.7, а для студента факультета C - 0.9. Студент сдал первую сессию. Какова вероятность, что он учится: a). на факультете A б). на факультете B в). на факультете C?

решать по формуле байеса 

$$ P(A|B) = \frac{P(A)*P(B|A)}{P(B)} $$

отношение вероятности того, что сдал студент выбранного факультета к вероятности того, что сдали все, т.е.
P(A) - вероятность того, что выбрали студента определенного факультета 

P(B|A) - априорная вероятность сдачи на этом факультеты

P(B) - вероятность сдачи всеми студентами

при этом, очевидно, что P(A) - определяется кол-вом студентов (A := 0.25, B := 0.25, C := 0.5)

```python
# вероятность, что сдали все
p_all = 0.25 * 0.8 + 0.25 * 0.7 + 0.5 * 0.9
# конкретно для факультетов
pa = (0.25 * 0.8 ) / p_all
pb = (0.25 * 0.7) / p_all
pc = (0.5 * 0.9) / p_all

print('а) вероятность сдачи студентом факультета A: {:.2%}'.format(pa))
print('б) вероятность сдачи студентом факультета B {:.2%}'.format(pb))
print('в) вероятность сдачи студентом факультета C: {:.2%}'.format(pc))
```

Устройство состоит из трех деталей. Для первой детали вероятность выйти из строя в первый месяц равна 0.1, для второй - 0.2, для третьей - 0.25. Какова вероятность того, что в первый месяц выйдут из строя: а). все детали б). только две детали в). хотя бы одна деталь г). от одной до двух деталей?

```python
p1 = 0.1
p2 = 0.2
p3 = 0.25

# все детали
p_all = p1 * p2 * p3

# только 2 - сумма вероятностей одновременных событий, когда выходят 2 детали, одна остается в строе
p_two = p1 * p2 * (1 - p3) + p1 * (1 - p2) * p3 + (1 - p1) * p2 * p3

# хотя бы одна деталь - это вероятность обратная тому, что все детали не выйдут из строя
p_lst = 1 - (1 - p1) * (1 - p2) * (1 - p3)

# от одной до двух - вероятность, что выйдут из строя как минимум одна минус вероятность, что выйдут все
p_one_two = p_lst - p_all

print('а) вероятность выйти из строя для всех деталей: {:.2%}'.format(p_all))
print('б) вероятность выйти из строя для только двух деталей: {:.2%}'.format(p_two))
print('в) вероятность выйти из строя для хотя бы одной детали: {:.2%}'.format(p_lst))
print('г) вероятность выйти из строя для от одной до двух деталей: {:.2%}'.format(p_one_two))
```
